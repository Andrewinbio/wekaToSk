```python
import pandas as pd
import numpy as np
```

```python
ori_data = pd.read_table('~/Downloads/2017-10-string-net-uniprot-full-links-v10.5-150.tsv',sep='\t')
ori_data.iloc[:10,:]
```

```python
features = list(ori_data)
features = features[2:]
features
```

```python
for value in features:
    selected_data = ori_data[['uniprot1','uniprot2',value]]
    selected_data[value] = selected_data[value].astype(int)
    selected_data = selected_data.drop_duplicates()
    df = selected_data.pivot_table(index='uniprot1',columns='uniprot2',values=value,aggfunc=np.max)
    df = df.fillna(0)
    df = df.astype(int)
    df = df.loc[:, (df != 0).any(axis=0)]
    df.to_csv('~/Desktop/StringData/stringdata/%s.csv' %value)
    print 'Shape of data: %d rows, %d columns. Done generating feature matrix %s' %(df.shape[0],df.shape[1],value)
```
